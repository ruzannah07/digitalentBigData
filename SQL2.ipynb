## One to many relationship

```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    password=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
#1. Report the account representative for each customer.

run("""
SELECT customerName, CONCAT(firstName,' ',lastName) AS employeeName
FROM customers JOIN employees
ON customers.salesRepEmployeeNumber = employees.employeeNumber
""")
 
```

```python
#2. Report total payments for Atelier graphique.
run("""
SELECT customerName, orderNumber, 
SUM(quantityOrdered*priceEach)AS totalBelanja
FROM customers 
JOIN orders USING(customerNumber) 
JOIN orderdetails USING (orderNumber)
GROUP BY orderNumber
""")
```

```python
#3. Report the total payments by date
run("""
SELECT orderDate, SUM(quantityOrdered*priceEach) AS totalPayments
FROM orders JOIN orderdetails USING (orderNumber)
GROUP BY orderDate
""")
```

```python
#4. Report the products that have not been sold
run("""
SELECT productName, shippedDate, status
FROM products JOIN orderDetails USING (productCode)
JOIN orders USING (orderNumber) 
WHERE shippedDate IS NULL
""")

#sama aja?

# run("""
# SELECT productName, shippedDate
# FROM orders JOIN orderDetails USING (orderNumber)
# JOIN products USING (productCode) 
# WHERE shippedDate IS NOT NULL AND status ='Cancelled'
# """)
```

```python
#5. List the amount paid by each customer.
run("""
SELECT customerName, SUM(amount) AS totalPayment
FROM payments LEFT JOIN customers USING (customerNumber)
GROUP BY customerNumber
""")
```

```python
#6. How many orders have been placed by Herkku Gifts?
run("""
SELECT COUNT(*)
FROM orders JOIN customers USING (customerNumber)
WHERE customerName = 'Herkku Gifts'
GROUP BY customerNumber
""")
```

```python
#7. Who are the employees in Boston?
run("""
SELECT CONCAT(firstName, ' ', lastName) AS employeeName
FROM employees JOIN offices USING (officeCode)
WHERE city ='Boston'
""")
```

```python
#8. Report those payments greater than \$100,000. 
#Sort the report so the customer who made the 
#highest payment appears first.

#TANYA: amount of payments = SUM(quantityOrderd*priceEach) 
#group by order number OF orderdetails
run("""
SELECT customerName, amount
FROM payments JOIN customers USING (customerNumber)
WHERE amount >100000
ORDER BY amount DESC
""")
```

```python
#9. List the value of 'On Hold' orders.
#10. Report the number of orders 'On Hold' for each customer.

run("""
SELECT customerName,orderNumber,SUM(quantityOrdered*priceEach)AS totalPaymentOnhold, status
FROM orders JOIN orderdetails USING (orderNumber)
JOIN customers USING (customerNumber)
WHERE orders.status='On hold'
GROUP BY orderNumber
""")
```

## Many to many relationship

```python
#1. List products sold by order date.
run("""
SELECT productName, orderDate
FROM products JOIN orderDetails USING (productCode)
JOIN orders USING (orderNumber)
WHERE shippedDate IS NOT NULL
""")

#tanya, product yg terjual?
#yg sudah dilakukan payment/ yg status != Canceled??
```

```python
#3. List the names of customers and their corresponding order number where
#a particular order from that customer has a value greater than $25,000?
run("""
SELECT customerName, SUM(quantityOrdered*priceEach) AS total
FROM customers JOIN orders USING (customerNumber)
JOIN orderdetails USING (orderNumber)
GROUP BY orderNumber
HAVING total > 25000
""")
```

```python
#4. Are there any products that appear on all orders?
run("""
SELECT productCode, COUNT(*) AS jmlKemunculan
FROM orderdetails
GROUP BY orderNumber
HAVING jmlKemunculan = (SELECT COUNT(*)FROM orders)
""")
```

```python
#5. List the names of products sold at less than 80% of the MSRP.
run("""
SELECT productName, priceEach, MSRP
FROM orderdetails JOIN products USING (productCode)
WHERE orderdetails.priceEach < (0.8*products.MSRP)
""")
```

```python
#6. Reports those products that have been sold with a markup of 100% 
#or more (i.e., the priceEach is at least twice the buyPrice)
run("""
SELECT productName, priceEach - buyPrice AS keuntungan, buyPrice
FROM products JOIN orderdetails
USING (productCode)
HAVING keuntungan > buyPrice
""")

#==== sama aja ======  
#WHERE : data murni, 
#HAVING: untuk kolom baru hasil pemrosesan data
run("""
SELECT productName, priceEach - buyPrice AS keuntungan, buyPrice
FROM products JOIN orderdetails
USING (productCode)
WHERE priceEach-buyPrice > buyPrice
""")
```

```python
#7. List the products ordered on a Monday.

run("""
SELECT productName
FROM products JOIN orderdetails USING (productCode)
JOIN orders USING (orderNumber)
WHERE DAYOFWEEK(orderDate) = 2 #1:minggu, 2:senin
""")
```

```python
#8. What is the quantity on hand for products listed on 'On Hold' orders?
#brp jumlah produk yg berstatus on hold
run("""
SELECT productCode, status
FROM orderdetails JOIN orders USING (orderNumber)
WHERE status = 'on hold'
""")
```
