```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = '3306'
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user = user,
    password = password,
    db = database,
    charset = 'utf8mb4')

def run(sql):
    df=pd.read_sql_query(sql,conn)
    return df
```

### 1. List products sold by order date.

```python
#SOLD : order yg telah dilakukan pembayaran
```

```python
run("""
select productname, orderdate
from products 
join orderdetails using (productcode)
join orders using (ordernumber)
join customers using (customerNumber)
join payments using (customerNumber)
""")
#menggunakan inner join, otomatis productname yg mempunyai relasi hingga ke table payments
```

### 2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck.

```python
run("""
select productName, orderdate
from orders 
join orderdetails using (ordernumber)
join products using (productcode)
where productName = '1940 Ford Pickup Truck' 
order by orderdate desc
""")
```

### 3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?


```python
run("""
select customername, ordernumber , sum(quantityordered*priceeach) as total
from customers join orders using (customernumber)
join orderdetails using (ordernumber)
group by ordernumber
having total > 25000
""")
```

### 5. List the names of products sold at less than 80% of the MSRP.


```python
run("""
select productname, priceEach, MSRP
from products join orderDetails using (productcode)
where priceEach < 0.8*msrp
""")
```

### 6. Reports those products that have been sold with a markup of 100% or more (i.e.,  the priceEach is at least twice the buyPrice)
 

```python
run("""
SELECT productName, priceEach - buyPrice AS keuntungan, buyPrice
FROM products JOIN orderdetails
USING (productCode)
HAVING keuntungan > buyPrice
""")
```

### 7. List the products ordered on a Monday.


```python
run("""
SELECT productName
FROM products JOIN orderdetails USING (productCode)
JOIN orders USING (orderNumber)
WHERE DAYOFWEEK(orderDate) = 2 #1:minggu, 2:senin
""")
```

### 8. What is the quantity on hand for products listed on 'On Hold' orders?

```python
run("""
SELECT productCode, status
FROM orderdetails JOIN orders USING (orderNumber)
WHERE status = 'on hold'
""")
```
