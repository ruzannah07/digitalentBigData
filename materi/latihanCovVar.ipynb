### ANALISIS DISTRIBUSI NORMAL

```python
m2 = np.mean(iris.data[:,0:2], axis=0)
m2
```

```python
#menghitung varian
v = np.var(iris.data[:,0])
v
```

```python
#menghitung covarian
C = np.cov(iris.data[:,0:2].T)
C
```

#### nilai covarian
    [[varian,covarian]
    [covarian,varian]]
    
    cov = 0, jika tidak ada hubungan antar var
      (-) negative, jika hubungan terbalik

```python
m1 = np.mean(iris.data[0:50, 0:2], axis = 0)
c1 = np.cov(iris.data[0:50, 0:2].T)

m2 = np.mean(iris.data[50:100, 0:2], axis = 0)
c2 = np.cov(iris.data[50:100, 0:2].T)

m3 = np.mean(iris.data[100:150, 0:2], axis = 0)
c3 = np.cov(iris.data[100:150, 0:2].T)

print(m1)
print(c1)
print("--------------------------")
print(m2)
print(c2)
print("--------------------------")
print(m3)
print(c3)
```

```python
#menentukan kemiringan dengan eigen value ???
w1, v1 = np.linalg.eig(c1)
print('eigenvalue =', w1)
print('eigenvector =', v1)
```

```python
w2, v2 = np.linalg.eig(c2)
print('eigenvalue2 =', w2)
print('eigenvector2 =', v2)
```

```python
w3, v3 = np.linalg.eig(c3)
print('eigenvalue3 =', w3)
print('eigenvector3 =', v3)
```

```python
x = [7.0, 1.6]
from scipy.stats import multivariate_normal
y1 = multivariate_normal.pdf(x, mean=m1, cov=c1)
print(y1)
y2 = multivariate_normal.pdf(x, mean=m2, cov=c2)
print(y2)
y3 = multivariate_normal.pdf(x, mean=m3, cov=c3)
print(y3)
```
