```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = '3306'
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user = user,
    password = password,
    db = database,
    charset = 'utf8mb4')

def run(sql):
    df=pd.read_sql_query(sql,conn)
    return df
```

### 1. Prepare a list of offices sorted by country, state, city.

```python
run("select * from offices order by country, state, city")
```
