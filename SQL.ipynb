```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
#1. Prepare a list of offices sorted by country, state, city.
run('SELECT*FROM offices ORDER BY country, state, city')
```

```python
#2. How many employees are there in the company?
run('SELECT COUNT(*) AS banyakKaryawan FROM employees')
```

```python
#3. What is the total of payments received?
run("SELECT SUM(amount) AS amountPayments FROM payments")
```

```python
#4. List the product lines that contain 'Cars'.
run("SELECT*FROM productlines WHERE productline LIKE '%Cars%'")
```

```python
#5. Report total payments for October 28, 2004.
run("""
    SELECT SUM(amount) AS Payments_Oct2804 
    FROM payments 
    WHERE paymentDate = '2004-10-28'
""")
```

```python
#6. Report those payments greater than \$100,000.
run("SELECT*FROM payments WHERE amount>100000")
```

```python
#7. List the products in each product line.
run("SELECT productLine,productName FROM products ORDER BY productLine")
```

```python
#8. How many products in each product line?
run("""
SELECT productLine,COUNT(*) AS totalProducts 
FROM products GROUP BY productLine
""")
```

```python
#9. What is the minimum payment received?
run("SELECT MIN(amount) FROM payments ")
```

```python
#10. List all payments greater than twice the average payment.
run("""
SELECT * FROM payments 
WHERE amount>(SELECT AVG(amount)*2 AS rerata FROM payments)
""")
```

```python
#11. What is the average percentage markup of the MSRP on buyPrice?
run("SELECT AVG(MSRP-buyPrice) AS rerataKeuntungan FROM products")
```

```python
#12. How many distinct products does ClassicModels sell?
run("SELECT DISTINCT productCode FROM products")
```

```python
#13. Report the name and city of customers who don't have sales representatives?
run("""
SELECT customerName, city, salesRepEmployeeNumber 
FROM customers LEFT JOIN employees
ON customers.salesRepEmployeeNumber = employees.employeeNumber
WHERE salesRepEmployeeNumber IS NULL
""")
```

```python
# 14. What are the names of executives with VP or Manager in their title?
# Use the CONCAT function to combine the employee's first name and 
# last name into a single field for reporting.

run("""SELECT CONCAT(firstName, ' ', lastName) AS nama, jobTitle 
FROM employees
WHERE jobTitle LIKE '%VP%' OR jobTitle LIKE '%Manager%'""")
```

```python
#15. Which orders have a value greater than $5,000?
run("""
SELECT orderNumber, SUM(priceEach*quantityOrdered) AS total
FROM orderdetails
GROUP BY orderNumber
HAVING total>5000
""")
```
